```python
# Paquetes requeridos 

import numpy as np
import skfuzzy as sk
import matplotlib.pyplot as plt

#Definiendo los rangos de velocidad de 0 a 80

x = np.arange (30, 80, 0.1)

#Se define un array para la función miembro de tipo triangular

lento = sk.trimf(x, [30, 30, 50])
medio = sk.trimf(x, [30, 50, 70])
medio_rapido = sk.trimf(x, [50, 60, 80])
rapido = sk.trimf(x, [60, 80, 80])

#Dibujando las funciones membresia

plt.figure()
plt.plot(x, rapido, 'b', linewidth=1.5, label='Rapido')
plt.plot(x, medio_rapido, 'k', linewidth=1.5, label='Medio Rapido')
plt.plot(x, medio, 'y', linewidth=1.5, label='Medio')
plt.plot(x, lento, 'r', linewidth=1.5, label='Lento')
plt.title('Penalti difuso')
plt.ylabel('Membresia')
plt.xlabel('Velocidad (km/h)')
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1, fancybox=True, shadow=True)
plt.show()
```
